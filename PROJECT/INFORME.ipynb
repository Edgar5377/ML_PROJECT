```python
# Count the number of different elements in the column 'FUENTE_FINANC'
unique_values_count = result['ACT_OBRA_ACCINV'].value_counts()
# print(unique_values_count)
# Plot the count of unique elements
plt.figure(figsize=(10, 6))
unique_values_count.plot(kind='bar')
plt.title('Count of Unique Elements in FUENTE_FINANC')
plt.xlabel('FUENTE_FINANC')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## Plot all the countplots of every variable in the dataset

```python

# Set up the subplots grid
num_cols = len(result.columns)
num_rows = num_cols // 3 + (1 if num_cols % 3 != 0 else 0)

fig, axes = plt.subplots(num_rows, 3, figsize=(15, 5*num_rows))
axes = axes.flatten()

# Iterate through each column
for i, column in enumerate(result.columns):
    # Count the number of different elements in the column
    unique_values_count = result[column].value_counts()
    
    # Plot the count of unique elements
    unique_values_count.plot(kind='bar', ax=axes[i])
    axes[i].set_title(f'Count of Unique Elements in {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=45)

# Hide empty subplots if there are any
for i in range(len(result.columns), len(axes)):
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```
